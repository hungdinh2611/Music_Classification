```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_pp,y_pp,test_size=0.2,random_state=42)
```

## Model selection

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
models = {'rfc': {
                'model': RandomForestClassifier(n_jobs=1),
                'params' : { 
                    'criterion' : ['gini','entropy'],
                    'min_samples_split': [2,3,4]
                }
            },
          'knn' : { 
                    'model' : KNeighborsClassifier(n_jobs=1),
                    'params':{
                        'n_neighbors' : [4,5,6],
                        'weights' : ['uniform','distance']
                   }
            },
          'lrc' : {
                  'model' : LogisticRegression(solver='saga',n_jobs=1,random_state=42),
                  'params' : {
                      'C' : [1,5,10]
                  }
          }
}
```

```python
model_scores = []
for model,mp in models.items():
    clf=GridSearchCV(mp['model'],mp['params'],verbose=4)
    clf.fit(X_pp,y_pp)
    model_scores.append({
        'model' : model,
        'params' : clf.best_params_,
        'score' : clf.best_score_
    })
    
    
```

```python
model_scores
```

```python
rfc_model= RandomForestClassifier(criterion='entropy',
                                  min_samples_split=2)
rfc_model.fit(X_train,y_train)

knn_model= KNeighborsClassifier(n_neighbors=6,
                               weights='distance')
knn_model.fit(X_train,y_train)

'''lrc_model = LogisticRegression(C=5)
lrc_model.fit(X_train,y_train)''';
```

```python
from sklearn.metrics import accuracy_score
y_pred = knn_model.predict(X_test)
accuracy_score(y_pred,y_test)
```
